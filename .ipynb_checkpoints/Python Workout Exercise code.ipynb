```python
# Exercise 1 Number Guessing Game (p.2)

import random

def guessing_game():
        
    rand_int = random.randint(1, 100)
    
    while True: 
      
        user_input = int(input("Please enter a number between 1 and 100: "))
        
        if user_input > rand_int:
            print("Too high! Guess again")
            continue
        elif user_input < rand_int:
            print("Too low! Guess again!")
            continue
        else:
            print("Just right!")
            break

guessing_game()
```

```python
# Beyond the exercise (p.10)

"""Write a function that takes a list of words (strings). It should return a tuple containing
three integers, representing the length of the shortest word, the length of the longest word, 
and the average word length."""

def tuple_with_lenghts(list_of_words):
    
    mylist = []
    
    max_word = max(list_of_words, key=len)
    min_word = min(list_of_words, key=len)
    avg_word = sum(len(i) for i in list_of_words) / len(list_of_words)
    
    mylist = [max_word, min_word, avg_word]
    
    print(tuple(mylist))
    

tuple_with_lenghts(["four", "fives", "two"])
```

```python
# Beyond the exercise (p.10)

"""Write a function that takes a list of Python objects. Sum the objects that either
are integers or can be turned into integers, ignoring the others."""

def sum_objects(my_list):
    
    new_list = []
    
    for i in my_list:
        if type(i) == int:
            new_list.append(i)
        elif i.isnumeric():
            new_list.append(int(i))
        else:
            continue
    
    return sum(new_list)

sum_objects([1,2,3, "3", "a"])
```

```python
#Exercise 3 Run Timing (p.11) 

"""Write a function (run_timing) that asks how long it took for you to run 10 km.
The function continues to ask how long (in minutes) it took for additional runs, until
the user presses Enter. At that point, the function exits—but only after calculating and
displaying the average time that the 10 km runs took."""

total = 0
count = 0
average = 0

while True:
    text = input("Enter 10 km running time (or press Enter to Exit): ")
    if text == "":
        break
    text = float(text)
    total += text
    count +=1
    average = total / count
    
print(f"Average of {average}, over {count} runs.")
```

```python
# Beyond the exercise (p.13)

"""Write a function that takes a float and two integers (before and after). The
function should return a float consisting of before digits before the decimal
point and after digits after. Thus, if we call the function with 1234.5678, 2 and
3, the return value should be 34.567."""

def edit_float(my_float, before, after):
    
    # convert float to string to be able to edit it
    my_float = str(my_float)
    
    # split string into two parts so you can edit them separately
    split_string = my_float.split(".")
    
    # string slicing including before and after variables
    left_string = split_string[0][-before:]
    right_string = split_string[1][0:after]
    
    # the f-string combines the two strings with a dot in between
    print(f"{left_string}.{right_string}")
    
edit_float(1234.5678, 2, 3)
```

```python
# Beyond the exercise (p.21)
# Pig latin function handling punctuation;  interpunction from input is moved to the end of the output

import string

def pig_latin_punc(word):
    
    #define variable with all vowels
    vowels = 'aeiou'

    # store string with all punctuation in a variable
    punctuation = string.punctuation
    
    if word[0] in vowels:        
        pig_latin = word + "way"
        
        if word[-1] in punctuation:            
            pig_latin = word[0:-1] + "way" + word[-1]
        
    else:
        
        if word[-1] in punctuation:            
             pig_latin = word[1:-1] + word[0] + "ay" + word[-1]
                
        else:
             pig_latin = word[1:] + word[0] + "ay"                
    
    print(pig_latin)
```

```python
# Beyond the exercise (p.21)

"""Consider an alternative version of Pig Latin—We don’t check to see if the first letter
is a vowel, but, rather, we check to see if the word contains two different vowels.
If it does, we don’t move the first letter to the end. Because the word “wine”
contains two different vowels (“i” and “e”), we’ll add “way” to the end of it, giving
us “wineway.” By contrast, the word “wind” contains only one vowel, so we
would move the first letter to the end and add “ay,” rendering it “indway.” How
would you check for two different vowels in the word? (Hint: sets can come in
handy here.)"""

def pig_latin_alt(word):
    
    if len(set(word) & set('aeiou')) > 1:
        
        word = word + "way"
        
        return word
        
    elif len(set(word) & set('aeiou')) == 1:
        
        word = word[1:] + "ay"
        
        return word
        
    else:
        print("no vowels found.")
    
    
pig_latin_alt("wine")        
```

```python
# CH2 Exercise 6 Pig latin Sentence (p.22)

"""Write a function called pl_sentence that takes a string containing several words, separated by
spaces. (To make things easier, we won’t actually ask for a real sentence. More specifically,
there will be no capital letters or punctuation.) So, if someone were to call pl_sentence('this is a test translation')
the output would be "histay isway away estay ranslationtay" 
Print the output on a single line, rather than with each word on a separate line."""

def pl_sentence(myline):
    
    vowels = "aeiou"
    
    mylist = []
    
    for word in myline.split(" "):
        if word[0] in vowels:
            word = word + "way"
            mylist.append(word)
            
        else:
            word = word[1:] + word[0] + "ay"
            mylist.append(word)
        
    output = " ".join(mylist)  
    
    return output

pl_sentence('this is a test translation')
```

```python
# Beyond the exercise (p.24)

"""Write a function that transposes a list of strings, in which each string contains
multiple words separated by whitespace. Specifically, it should perform in such a
way that if you were to pass the list ['abc def ghi', 'jkl mno pqr', 'stu vwx yz']
to the function, it would return ['abc jkl stu', 'def mno vwx', 'ghi pqr yz']."""

def transpose(a_string):
    
    new_string1 = (a_string[0].split(" ")[0] + " " + a_string[1].split(" ")[0] + " " + a_string[2].split(" ")[0])
    
    new_string2 = (a_string[0].split(" ")[1] + " " + a_string[1].split(" ")[1] + " " + a_string[2].split(" ")[1])
    
    new_string3 = (a_string[0].split(" ")[2] + " " + a_string[1].split(" ")[2] + " " + a_string[2].split(" ")[2])
    
    return [new_string1, new_string2, new_string3]

transpose(['abc def ghi', 'jkl mno pqr', 'stu vwx yz'])
```

```python
# Exercise 7 Ubbi Dubbi (p.24)

"""For this exercise, you’ll write a function (called ubbi_dubbi) that takes a single word (string) as an argument. 
It returns a string, the word’s translation into Ubbi Dubbi. So if the function is called with octopus, the function 
will return the string uboctubopubus. And if the user passes the argument elephant, you’ll output ubelubephubant."""

def ubbi_dubbi(word):
    
    vowels = 'aeiou'
    
    ubbi_du = [] 
    
    for letter in word:        
                        
        if letter in vowels:
            mystring = 'ub' + letter
            ubbi_du.append(mystring)
        else:
            ubbi_du.append(letter)
        
        
    return "".join(ubbi_du)
    
    
ubbi_dubbi('elephant')
```

```python
# Beyond the exercise (p.26)

""" In URLs, we often replace special and nonprintable characters with a % followed by the character’s ASCII value
in hexadecimal. For example, if a URL is to include a space character (ASCII 32, aka 0x20), we replace it with %20. 
Given a string, URL-encode any character that isn’t a letter or number.

reminder: hex() takes in int and converts it to its corresponding hexadecimal form, 
ord() returns the number representing the unicode code of a specified character.

"""

def sel_string():

    my_string = "ABC999zz#~€&/()=?¿"

    sel_string = []

    for char in my_string:
        # check if char is number
        if char.isdigit() == True:
            continue
        # check if char is letter
        elif char.isalpha() == True:
            continue
        # add char to empty list
        else:
            sel_string.append(char)
            
    # convert char in list to ASCII value in hexadecimal
    for i in sel_string:
        print(f"url encode of char {i} is: {hex(ord(i))}")
        

sel_string()
```

```python
# Ch3 Exercise "First Last" (p.31)

"""Write a function, firstlast, that takes a sequence (string, list, or tuple) and returns 
the first and last elements of that sequence, in a two-element sequence of the same type. 
So firstlast('abc') will return the string ac, while firstlast([1,2,3,4]) will return the list [1,4]."""

def firstlast(my_seq):
    
    if type(my_seq) == list or type(my_seq) == tuple:
        my_seq = list(my_seq)
        return [my_seq[0], my_seq[-1]]
    
    elif type(my_seq) == str:
        return my_seq[0] + my_seq[-1] 
        
   
firstlast((1,2,3,4))
```

```python
# Beyond the exercise (p.35)

"""Write a function that takes a list or tuple of numbers. Return a two-element list, containing (respectively) the sum
of the even-indexed numbers and the sum of the odd-indexed numbers. So calling the function as even_odd_sums([10, 20,
30, 40, 50, 60]), you’ll get back [90, 120]."""

def my_sum(my_seq):
    
    a, b = 0, 0
 
    for i in enumerate(my_seq):
        if i[0] % 2 == 0:
            a += i[1]
        else:
            b += i[1]
            
    return [a, b]
        

my_sum([10, 20, 30, 40, 50, 60])
```

```python
# Beyond the exercise (p.35)

"""Write a function that takes a list or tuple of numbers. Return the result of alternately adding and subtracting numbers
from each other. So calling the function as plus_minus([10, 20, 30, 40, 50, 60]), you’ll get back the result of
10+20-30+40-50+60, or 50."""

def plus_minus(myseq):
    
    total = myseq[0]
     
    for i in enumerate(myseq[1:]):
             
        if i[0] % 2 == 0:
            total += i[1]
        else:
            total -= i[1]
    
            
    return total            
        

plus_minus([10, 20, 30, 40, 50, 60])
```

```python
# Beyond the exercise (p.40)

"""Write a function, sum_numeric, that takes any number of arguments. If the
argument is or can be turned into an integer, then it should be added to the
total. Arguments that can’t be handled as integers should be ignored. The
result is the sum of the numbers. Thus, sum_numeric(10, 20, 'a', '30',
'bcd') would return 60. Notice that even if the string 30 is an element in the
list, it’s converted into an integer and added to the total."""

def sum_numeric(*args):
    
    total = 0
    
    for arg in args:
        if type(arg) == int:
            total += arg
        elif arg.isnumeric():
            total += int(arg)
            
    return total

sum_numeric(10, 20, 'a', '30', 'bcd')
```

```python
# Beyond the exercise (p.40)

"""Write a function that takes a list of dicts and returns a single dict that combines
all of the keys and values. If a key appears in more than one argument, the
value should be a list containing all of the values from the arguments."""

def list_of_dicts(mylist):
        
    newdict = {}
    
    for item in mylist:
        for k, v in item.items():
            
            if k in newdict:
                
                newdict[k].append(v)
                
            else:
                
                newdict[k] = [v]           
                
    return newdict
        
        
list_of_dicts([{'a': 1, 'b': 2}, {'b': 1, 'd': 2}])
```

```python
# Exercise 12 "Word with most repeated letters" (p.46)

"""Write a function, most_repeating_word, that takes a sequence of strings as input. The function should return the 
string that contains the greatest number of repeated letters. In other words:
*For each word, find the letter that appears the most times.
*Find the word whose most-repeated letter appears more than any other.
That is, if words is set to words = ['this', 'is', 'an', 'elementary', 'test', 'example'] then your function should 
return elementary."""

def most_repeating_letter_count(myseq):
    
    return sorted(myseq, key=lambda word: sum(ch in 'abcdefghijklmnopqrstuvwxyz' for ch in word), reverse=True)[0]

most_repeating_letter_count(['this', 'is', 'an', 'elementary', 'test', 'example'])
```

```python
# same function, with less code

from collections import Counter

WORDS = ['this', 'is', 'an', 'elementary', 'test', 'example']

def most_repeating_letter_count(word):
    
    return Counter(word).most_common(1)[0][1]

def most_repeating_word(words):
    
    return max(words, key=most_repeating_letter_count)


print(most_repeating_word(WORDS))
```

```python
# Beyond the exercise (p.49)

"""Instead of finding the word with the greatest number of repeated letters, find
the word with the greatest number of repeated vowels."""

WORDS = ['this', 'is', 'an', 'elementary', 'test', 'example', 'couscouscouscous']

def by_vowel_count(one_word):
    total = 0
    for one_character in one_word.lower():
        if one_character in 'aeiou':
            total += 1
    return total


def sort_by_vowel_count(words):
    """Given a list of strings (words), return
a list of those words sorted by the number of vowels
they contain.
"""
    return max(words, key=by_vowel_count)

print(sort_by_vowel_count(WORDS))
```

```python
# Exercise 13 "Printing tuple records" (p.49)

"""For this exercise, write a Python function, format_sort_records, that takes the
PEOPLE list and returns a formatted string that looks like the following:

Trump Donald 7.85
Putin Vladimir 3.63
Xi Jinping 10.60 """

PEOPLE = [('Donald', 'Trump', 7.85),
('Vladimir', 'Putin', 3.626),
('Jinping', 'Xi', 10.603)]

def format_sort_records(people):
    
    for person in sorted(people, key=lambda x: x[0]):
        print(f"{person[1]: <10} {person[0]: <10} {person[2]: .2f}")
        
format_sort_records(PEOPLE)
```

```python
# Beyond the exercise (p.51)

from operator import attrgetter
from collections import namedtuple

Person = namedtuple('Person', ['first', 'last', 'distance'])


PEOPLE = [Person('Donald', 'Trump', 7.85),
          Person('Vladimir', 'Putin', 3.626),
          Person('Jinping', 'Xi', 10.603)]


def format_sort_records(list_of_tuples):
    output = []
    
    for person in sorted(list_of_tuples, key=attrgetter('first')):
        output.append(person._asdict())
        
    for person in output:
        print(f"{person['last']: <10} {person['first']: <10} {person['distance']: .2f}")
        
    
format_sort_records(PEOPLE)
```

```python
# Exercise 14 "Restaurant" (p.57)

"""In this exercise, I want you to create a new constant dict, called MENU, representing
the possible items you can order at a restaurant. The keys will be strings, and the values
will be prices (i.e., integers). You should then write a function, restaurant, that
asks the user to enter an order:
* If the user enters the name of a dish on the menu, the program prints the price
and the running total. It then asks the user again for their order.
* If the user enters the name of a dish not on the menu, the program scolds the
user (mildly). It then asks the user again for their order.
* If the user enters an empty string, the program stops prompting and prints the
total amount."""



def restaurant():
    
    total = 0
    
    while True:
    
        MENU = {'wine': 10, 'sandwitch': 12, "burger": 14, "steak": 16}

        order = input("please enter your order: ")            

        if order == "":
            
            break
        
        elif order in MENU.keys():
            price = MENU[order]
            total+= price
            print(f"price of {order} is {price}, total is {total}")
            order

        elif order not in MENU:
            print(f"sorry, we are fresh out of {order} today")
            order
        
        
    print(f"total is {total}")  

restaurant()
```

```python
# Beyond the exercise (p.58)

"""Define a dict whose keys are dates (represented by strings) from the most recent
week and whose values are temperatures. Ask the user to enter a date, and display
the temperature on that date, as well as the previous and subsequent dates,
if available."""

temperatures = {"3-10-2022": 32, "4-10-2022": 30, "5-10-2022": 33, "6-10-2022": 31, "7-10-2022": 32}

def print_temp():
    
    date = input("please enter a date: ")
    
    new_date = date.split("-")
    
    sub_day, pre_day = str(int(new_date[0]) + 1), str(int(new_date[0]) - 1)
    
    template = f"{new_date[0]}-{new_date[1]}-{new_date[2]}"
    
    sub_temp, pre_temp = f"{sub_day}-{new_date[1]}-{new_date[2]}", f"{pre_day}-{new_date[1]}-{new_date[2]}"
    
    if template in temperatures:
        
        print(f"Temperature on {date} was {temperatures[date]} degrees.") 
        
    if sub_temp in temperatures:        
        
        print(f"Temperature on {sub_temp} was {temperatures[sub_temp]} degrees.")
        
    if pre_temp in temperatures:
            
        print(f"Temperature on {pre_temp} was {temperatures[pre_temp]} degrees.")
            
    else:
        print("No date found.")
    
    
print_temp()           
```

```python
# Beyond the exercise (p.59)

"""Define a dict whose keys are names of people in your family, and whose values
are their birth dates, as represented by Python date objects (http://mng.bz/
jggr). Ask the user to enter the name of someone in your family, and have the
program calculate how many days old that person is."""

from datetime import date, timedelta

def calc_days_since_birth():    

    family_birthdays = {"eric": date(1980, 2, 5), 'luuk': date(1982, 7, 27), 'rene': date(1984, 3, 27)}

    today = date.today()

    family_name = input("please enter family name: ")

    if family_name in family_birthdays:        
    
        t = today - family_birthdays[family_name]

        print(f"family member {family_name} is {t} days old.")
        
    else:
        
        print("No such name in database.")

    
calc_days_since_birth()
```

```python
# Exercise 15 "Rainfall" (p.59)

"""Specifically, write a function, get_rainfall, that tracks rainfall in a number of cities.
Users of your program will enter the name of a city; if the city name is blank, then
the function prints a report (which I’ll describe) before exiting.
If the city name isn’t blank, then the program should also ask the user how much
rain has fallen in that city (typically measured in millimeters). After the user enters
the quantity of rain, the program again asks them for a city name, rainfall amount,
and so on—until the user presses Enter instead of typing the name of a city."""

def get_rainfall():
    
    city_rain = {}
    
    while True:
        
        city = input("Enter the name of a city: ")
        
        if not city:
            break
        
        rain = input("How much rain has fallen in the the same city?")

        city_rain[city] = city_rain.get(city, 0) + int(rain)
        
            
    for key, value in city_rain.items():
        print(f"{key} \n{value}")
            
get_rainfall()   
```

```python
# Beyond the exercise (p.64)

"""Read through a text file on disk. Use a dict to track how many words of each
length are in the file—that is, how many three-letter words, four-letter words,
five-letter words, and so on. Display your results."""

def word_length():    

    # input text file
    inputFile = r"C:\Users\mipc\Downloads\words.txt"

    # define empty dict
    word_length_dict = {}

    # Opening the given file in read-only mode.
    with open(inputFile, 'r') as filedata:

        # Getting the list of words of a file
        wordsList = filedata.read().split()

        # fill dict with word as key and word length as value
        for word in wordsList:
            word_length_dict[word] = len(word)

    return word_length_dict

word_length()
```

```python
# Exercise 16 "Dictdiff" (p.64)

"""Write a function, dictdiff, that takes two dicts as arguments. The function returns
a new dict that expresses the difference between the two dicts.
"""

def dictdiff(a, b):
    
    non_common_pairs = dict()
    
    for key in a:        
        if (key not in b):            
            non_common_pairs[key] = [None, a[key]]
        
        elif (key in b):            
            if a[key] != b[key]:                
                non_common_pairs[key] = [a[key], b[key]] 
    
    for key in b:        
        if (key not in a):            
            non_common_pairs[key] = [None, b[key]]
            
    return non_common_pairs
            

d1 = {'a':1, 'b':2, 'c':3}       
d5 = {'a':1, 'b':2, 'd':4}
print(dictdiff(d1, d5))
```

```python
# same function, now using set.union()

def dictdiff(first, second):
    
    output = {}
    all_keys = set()
    all_keys = all_keys.union(first.keys(), second.keys())

    for key in all_keys:
        if first.get(key) != second.get(key):
            output[key] = [first.get(key), second.get(key)]
    return output

d1 = {'a':1, 'b':2, 'c':3}       
d5 = {'a':1, 'b':2, 'd':4}
print(dictdiff(d1, d5))
```
